```python
# Convert 'Year' to integer if necessary
movie['Year'] = movie['Year'].astype(int)

# Input year from the user
year = int(input("Enter the year: "))

# Initialize a list to store titles
titles = []

# Loop through the DataFrame and collect titles of movies released in the given year
for index, row in movie.iterrows():
    if row['Year'] == year:
        titles.append(row['Title'])

# Print the titles
print(f"Movies released in {year}:")
for title in titles:
    print(title)

# Print the total number of movies
print(f"Number of movies released in {year}: {len(titles)}")
```

```python
plt.title("movies vs years")
sb.barplot(x = 'Year', y = 'Title',data=movie.head())
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Year' to integer if necessary
movie['Year'] = movie['Year'].astype(int)

# Create a bar plot for the first 10 entries
sns.catplot(data=movie.head(10), x="Year", y="MetaScore", hue="Genre", kind="bar")
plt.title("Movies by Year and Genre")
plt.show()

```

```python
# Filter movies by year
year = int(input("Enter the year: "))
filtered_movies = movie[movie['Year'] == year]

# Count certificates
certificate_counts = filtered_movies['Certificates'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(certificate_counts, labels=certificate_counts.index, autopct='%1.1f%%')
plt.title(f"Certificates for Movies in {year}")
plt.show()
```

```python
# Count the number of movies each director made
director_counts = movie['Director'].value_counts()

for director, count in director_counts.items():
    print(f"{director}: {count} movies")
```

```python
plt.figure(figsize=(10, 6))
sb.countplot(data=movie, x='Year')
plt.title('Number of Movies Released Per Year')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sb.boxplot(data=movie, x='Genre', y='IMDb Rating')
plt.title('Distribution of IMDb Ratings by Genre')
plt.show()
```

```python
#Scatter Plot: Year vs. MetaScore
plt.figure(figsize=(10, 6))
sb.scatterplot(data=movie, x='Year', y='MetaScore', hue='Genre')
plt.title('MetaScore by Year')
plt.show()
```

```python
#Histogram: Distribution of MetaScores
plt.figure(figsize=(8, 5))
sns.distplot(movie['MetaScore'].dropna(), bins=20, kde=True)
plt.title('Distribution of MetaScores')
plt.xlabel('MetaScore')
plt.show()
```

```python
#Line Plot: Average MetaScore by Year
avg_metascore_by_year = movie.groupby('Year')['MetaScore'].mean().reset_index()
plt.figure(figsize=(10, 6))
sb.lineplot(data=avg_metascore_by_year, x='Year', y='MetaScore')
plt.title('Average MetaScore by Year')
plt.xticks(rotation=45)
plt.show()
```

```python
#Count Plot: Number of Movies by Genre
plt.figure(figsize=(12, 6))
sb.countplot(data=movie, x='Genre')
plt.title('Number of Movies by Genre')
plt.xticks(rotation=45)
plt.show()
```
