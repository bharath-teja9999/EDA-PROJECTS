```python
# Convert 'Year' to integer if necessary
movie['Year'] = movie['Year'].astype(int)

# Input year from the user
year = int(input("Enter the year: "))

# Initialize a list to store titles
titles = []

# Loop through the DataFrame and collect titles of movies released in the given year
for index, row in movie.iterrows():
    if row['Year'] == year:
        titles.append(row['Title'])

# Print the titles
print(f"Movies released in {year}:")
for title in titles:
    print(title)

# Print the total number of movies
print(f"Number of movies released in {year}: {len(titles)}")
```

```python
# Function to categorize movies based on duration
def categorize_movie(duration):
    if duration <= 80:
        return "short movie"
    elif 81 <= duration <= 120:
        return "medium time movie"
    else:
        return "big movie"

# Take input from the user for the movie name
movie_name = input("Enter the movie name: ")

# Find the movie in the dataset
movie_row = movie[movie['Title'].str.contains(movie_name, case=False, na=False)]

# Check if the movie is found
if not movie_row.empty:
    # Get the duration of the movie
    duration = movie_row['Duration (minutes)'].values[0]
    
    # Categorize the movie
    category = categorize_movie(duration)
    
    # Print the result
    print(f"The movie '{movie_name}' with duration {duration} minutes is considered a {category}.")
else:
    print(f"Movie '{movie_name}' not found in the dataset.")

    
```

```python
# Filter movies with duration less than 100 minutes
short_movies = movie[movie['Duration (minutes)'] < 100]

# Print the titles of the short movies
print("Movies with duration below 100 minutes:")
for title in short_movies['Title']:
    print(title)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Year' to integer if necessary
movie['Year'] = movie['Year'].astype(int)

# Create a bar plot for the first 10 entries
sns.catplot(data=movie.head(10), x="Year", y="MetaScore", hue="Genre", kind="bar")
plt.title("Movies by Year and Genre")
plt.show()

```

```python
# Filter movies by year
year = int(input("Enter the year: "))
filtered_movies = movie[movie['Year'] == year]

# Count certificates
certificate_counts = filtered_movies['Certificates'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(certificate_counts, labels=certificate_counts.index, autopct='%1.1f%%')
plt.title(f"Certificates for Movies in {year}")
plt.show()
```

### Dataset Overview:
The dataset contains information on 400 movies with attributes such as Title, IMDb Rating, Year, Certificates, Genre, Director, Star Cast, MetaScore, and Duration.The dataset spans a wide range of years from 1939 to 2025, indicating both historical and upcoming movie releases.

The average IMDb rating for the movies is 7.11, with a standard deviation of 0.93. This suggests that the majority of movies have above-average ratings.The average MetaScore is 66.90, indicating a generally positive critical reception.The average duration of the movies is approximately 116.33 minutes.
### Movies with High IMDb Ratings:
A significant number of movies have IMDb ratings above 7.0, indicating a predominance of well-received films in the dataset.

Movies are categorized based on duration: short movies (≤ 80 minutes), medium time movies (81-120 minutes), and big movies (> 120 minutes). For example, "Mad Max" and "Shrek" are short movies.
