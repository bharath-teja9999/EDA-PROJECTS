```python
sb.barplot(x ='Gender',y = 'Age',data = a)
```

```python
b = a['GroupSize'].head(10)
c = a['VisitFrequency'].head(10)
plt.pie(b,labels = c,autopct = '%1.1f%%',startangle = 50)
plt.show()
```

```python
#Distribution of Average Spend
plt.figure(figsize=(8, 5))
sb.distplot(a['AverageSpend'], hist=True,kde=True  )
plt.title('Distribution of Average Spend')
plt.xlabel('Average Spend')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Average Spend
plt.figure(figsize=(8, 5))
sb.distplot(a['AverageSpend'],hist=True,kde=False  )
plt.title('Distribution of Average Spend')
plt.xlabel('Average Spend')
plt.ylabel('Frequency')
plt.show()
```

```python
#Distribution of Income
plt.figure(figsize=(8, 5))
sb.distplot(a['AverageSpend'], hist=True, kde=True)
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
#Count of Preferred Cuisine
plt.figure(figsize=(8, 5))
sb.countplot(x='PreferredCuisine', data=a)
plt.title('Count of Preferred Cuisine')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()
```

```python
#Average Spend by Gender
plt.figure(figsize=(8, 5))
sb.boxplot(x='Gender', y='AverageSpend', data=a)
plt.title('Average Spend by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Spend')
plt.show()
```

```python
#Wait Time by Dining Occasion
plt.figure(figsize=(8, 5))
sb.violinplot(x='DiningOccasion', y='WaitTime', data=a)
plt.title('Wait Time by Dining Occasion')
plt.xlabel('Dining Occasion')
plt.ylabel('Wait Time')
plt.show()
```

```python
#Visit Frequency Count
plt.figure(figsize=(8, 5))
sb.countplot(x='VisitFrequency', data=a)
plt.title('Visit Frequency Count')
plt.xlabel('Visit Frequency')
plt.ylabel('Count')
plt.show()
```
