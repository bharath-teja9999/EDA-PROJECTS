```python
data['dteday'] = pd.to_datetime(data['dteday'])

# Define the date for comparison
comparison_date = pd.to_datetime('2012-06-20')

# Filter rows where 'dteday' is after the specified date
filtered_data = data[data['dteday'] > comparison_date]

# Print the filtered rows with separation
for i in range(len(filtered_data)):
    row = filtered_data.iloc[i]
    print(row[['dteday', 'weekday']])
    print('-' * 20)
```
